# Numero Primo

```python
def eh_primo(n):
    # Números menores que 2 não são primos.
    if n < 2:
        return False
    # 2 é o único número primo par.
    if n == 2:
        return True
    # Qualquer outro número par (maior que 2) não é primo.
    if n % 2 == 0:
        return False

    # Verifica divisores ímpares a partir de 3 até a raiz quadrada de n.
    # Não é necessário verificar além da raiz quadrada de n, pois se houver um divisor
    # maior que a raiz, haverá um menor também.
    # Incrementa de 2 em 2 para verificar apenas números ímpares.
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False

    # Se nenhum divisor for encontrado, o número é primo.
    return True
```

# Inversão de string

```python
def inverte_texto(texto):
    # Utiliza slicing de string para inverter o texto de forma concisa.
    # [::-1] cria uma cópia reversa da string.
    return texto[::-1]
```
