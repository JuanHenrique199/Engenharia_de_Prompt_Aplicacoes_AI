```python
# Initialize the sum to 0
soma = 0
# Define a list of numbers
numeros = [1,2,3,4,5,6]
# Iterate through each number in the list
for num in numeros:
  # Add the current number to the sum
  soma += num
# Print the total sum
print(soma)
```
