```python
def calculate_average(numbers):
    """
    Calcula a média de uma lista de números.

    Args:
        numbers (list): Uma lista de números (inteiros ou floats).

    Returns:
        float: A média dos números na lista.

    Raises:
        ValueError: Se a lista de entrada estiver vazia.
        TypeError: Se algum elemento da lista não for um número.
    """
    # Verifica se a lista está vazia para evitar divisão por zero
    if not numbers:
        raise ValueError("A lista de entrada não pode estar vazia. Por favor, forneça uma lista com pelo menos um número.")

    total_sum = 0
    count = 0

    # Itera sobre cada elemento da lista para somar e contar
    for num in numbers:
        # Tratamento rigoroso para garantir que todos os elementos sejam números
        if not isinstance(num, (int, float)):
            raise TypeError(f"Todos os elementos da lista devem ser números. Encontrado tipo inválido: {type(num).__name__}")
        total_sum += num
        count += 1

    # Calcula a média
    average = total_sum / count
    return average

# --- Casos de Teste ---

print("--- Iniciando Casos de Teste ---")

# Caso de Teste 1: Lista de números positivos
print("\nCaso de Teste 1: Lista com números positivos")
list1 = [10, 20, 30, 40, 50]
try:
    avg1 = calculate_average(list1)
    print(f"Lista: {list1} -> Média: {avg1}") # Saída esperada: 30.0
except (ValueError, TypeError) as e:
    print(f"Erro: {e}")

# Caso de Teste 2: Lista com números negativos e floats
print("\nCaso de Teste 2: Lista com números negativos e floats")
list2 = [-5, 0, 5.5, 10, -2.5]
try:
    avg2 = calculate_average(list2)
    print(f"Lista: {list2} -> Média: {avg2}") # Saída esperada: 1.6
except (ValueError, TypeError) as e:
    print(f"Erro: {e}")

# Caso de Teste 3: Lista vazia (espera-se um erro)
print("\nCaso de Teste 3: Lista vazia (espera-se um erro ValueError)")
list3 = []
try:
    avg3 = calculate_average(list3)
    print(f"Lista: {list3} -> Média: {avg3}")
except ValueError as e:
    print(f"Erro esperado capturado: {e}")
except TypeError as e:
    print(f"Erro inesperado de tipo capturado: {e}")

print("\n--- Casos de Teste Concluídos ---")
```
